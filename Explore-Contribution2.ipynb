```python
import math

def get_phases(input_list):
    output_list = list()
    for i in range(0,len(input_list)):
        phase = -1
        if input_list[i] >= 0 and input_list[i] <= 2:
            phase = "S"
        elif input_list[i] > 2 and input_list[i] <= 4.8:
            phase = "G2M"
        elif input_list[i] > 4.8 and input_list[i] <= 5.9:
            phase = "G1"
        elif input_list[i] > 5.9:
            phase = "S"
        output_list.append(phase)
    return(output_list)

names_for_all_time_courses_list = list()
time_course_1_angles_list = list()
time_course_2_angles_list = list()
time_course_3_angles_list = list()
time_course_4_angles_list = list()
time_course_1_times_list = list()
time_course_2_times_list = list()
time_course_3_times_list = list()
time_course_4_times_list = list()


# calculate the angles for the 4 time courses
for i in range(0,40,2): #t-course 1 T = 17.65, φ = 0.0
    #print(math.atan2(math.sin((i/17.65)*2*math.pi),math.cos((i/17.65)*2*math.pi))) # with negative radians
    time_course_1_angles_list.append(((i/17.65)*2*math.pi)%(2*math.pi)) # 0 to 2pi
    time_course_1_times_list.append(i)
    names_for_all_time_courses_list.append("t1_" + str(i))
for i in range(0,48,2): #t-course 2 T = 18.6, φ = 0.0
    #print(math.atan2(math.sin((i/18.6)*2*math.pi),math.cos((i/18.6)*2*math.pi))) # with negative radians
    time_course_2_angles_list.append(((i/18.6)*2*math.pi)%(2*math.pi)) # 0 to 2pi
    time_course_2_times_list.append(i)
    names_for_all_time_courses_list.append("t2_" + str(i))
for i in range(0,48,2): #t-course 3 T = 18, φ = 0.0
    #print(math.atan2(math.sin((i/18)*2*math.pi),math.cos((i/18)*2*math.pi))) # with negative radians
    time_course_3_angles_list.append(((i/18)*2*math.pi)%(2*math.pi)) # 0 to 2pi
    time_course_3_times_list.append(i)
    names_for_all_time_courses_list.append("t3_" + str(i))
for i in range(0,46,2): #t-course 4 T = 23.95, φ = 2.3
    #print(math.atan2(math.sin((i/23.95)*2*math.pi + 2.3),math.cos((i/23.95)*2*math.pi + 2.3))) # with negative radians
    time_course_4_angles_list.append(((i/23.95)*2*math.pi + 2.3)%(2*math.pi)) # 0 to 2pi
    time_course_4_times_list.append(i)
    names_for_all_time_courses_list.append("t4_" + str(i))

angles_list = time_course_1_angles_list + time_course_2_angles_list + time_course_3_angles_list + time_course_4_angles_list
whitfield_timepoints_phases = get_phases(angles_list)
time_course_1_phase_list = get_phases(time_course_1_angles_list)
time_course_2_phase_list = get_phases(time_course_2_angles_list)
time_course_3_phase_list = get_phases(time_course_3_angles_list)
time_course_4_phase_list = get_phases(time_course_4_angles_list)
print(time_course_1_phase_list)
print(time_course_2_phase_list)
print(time_course_3_phase_list)
print(time_course_4_phase_list)
```

```python
import matplotlib.pyplot as plt
plt.scatter(time_course_1_times_list, time_course_1_angles_list, c='red')
plt.scatter(time_course_2_times_list, time_course_2_angles_list, c='green')
plt.scatter(time_course_3_times_list, time_course_3_angles_list, c='blue')
plt.scatter(time_course_4_times_list, time_course_4_angles_list, c='orange')
plt.axhline(y=2, color='gray', linestyle='--')
plt.axhline(y=4.8, color='gray', linestyle='--')
plt.axhline(y=5.9, color='gray', linestyle='--')
plt.xlabel('Timepoints')
plt.ylabel('Angles: S, G2M, G1, S')
plt.title('Timecourse Angle Progression')
plt.show()
```
